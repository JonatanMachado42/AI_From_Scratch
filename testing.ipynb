```python
import numpy as np
```

```python

# Tamanho do dataset
n = 200
rng = np.random.default_rng(42)

# Features
x1 = rng.normal(0, 1, n)
x2 = rng.uniform(-2, 3, n)

# Target: y = 4*x1 - 2*x2 + ru√≠do
noise = rng.normal(0, 0.5, n)
y = 4*x1 - 2*x2 + noise

# Empilha em formato (n, 3)
dataset = np.column_stack((x1, x2, y))

print(dataset[:5])
# np.savetxt("dataset_regressao.csv", dataset, delimiter=",")
```
